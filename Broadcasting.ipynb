In Numpy, broadcasting is a mechanism that allows operations between arrays of different shapes. It allows you to perform mathematical operations on arrays with different shapes, by automatically replicating the smaller array to match the shape of the larger one

```python
import numpy as np

x = np.array([
    [3, 4, 5],
    [2, 6, 3],
    [8, 4, 1]
])

v = np.array([1, 0, 1])

print(x + v)
```

```python
print(x*v)
```
