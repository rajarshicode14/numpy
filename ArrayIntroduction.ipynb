```python
# The numpy.asarray() method
# This function is used to create an array by using the existing data in the form of lists, or tuples. 
# This function is useful in the scenario where we need to convert a python sequence into the numpy array object.

# Syntax: numpy.asarray(arr, [dtype], [order])
```

```python
# List to array
import numpy as np
my_list = [1, 3, 5, 7, 9]
  
print ("Input  list : ", my_list)
   
    
out_arr = np.asarray(my_list)
print ("output array from input list : ", out_arr) 
```

```python
# Tuple to array

my_tuple = ([1, 3, 9], [8, 2, 6])
   
print ("Input  tuple : ", my_tuple)
    
out_arr = np.asarray(my_tuple) 
print(out_arr)
```
